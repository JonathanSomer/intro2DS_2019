```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

# fix random seed for reproducibility
numpy.random.seed(7)
```

```python
import pandas as pd

df = pd.read_csv('pima-indians-diabetes.csv', header=None)
print(df.shape)
```

```python
df.head()
```

### column meanings
   1. Number of times pregnant
   2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
   3. Diastolic blood pressure (mm Hg)
   4. Triceps skin fold thickness (mm)
   5. 2-Hour serum insulin (mu U/ml)
   6. Body mass index (weight in kg/(height in m)^2)
   7. Diabetes pedigree function
   8. Age (years)
   9. Class variable (0 or 1)

```python
X = df.iloc[:,0:8]
Y = df.iloc[:,8]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=150, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
predictions = model.predict(X_test)
print(predictions[0:10])
```

```python
# round predictions
y_pred = [round(x[0]) for x in predictions]
print(y_pred[0:10])
```

```python
print(classification_report(y_test, y_pred))
```

### is this better than good old RandomForest?

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=150, max_depth=3,random_state=0)
clf.fit(X_train, y_train)  
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

### what if we add a bit of regularization?

```python
from keras.layers import Dropout
# create model
model_2 = Sequential()
model_2.add(Dense(12, input_dim=8, activation='relu'))
model_2.add(Dropout(0.1, input_shape=(12,)))
model_2.add(Dense(8, activation='relu'))
model_2.add(Dropout(0.1, input_shape=(8,)))
model_2.add(Dense(1, activation='sigmoid'))
model_2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_2.fit(X_train, y_train, epochs=150, batch_size=10)
predictions = model_2.predict(X_test)
y_pred = [round(x[0]) for x in predictions]
```

```python
print(classification_report(y_test, y_pred))
```
